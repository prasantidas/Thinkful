# 3
The prime factors of 13195 are 5, 7, 13 and 29.

What is the largest prime factor of the number 600851475143 ?

```python
def prime_factors(n):
    i = 2
    factors = []
    while i * i <= n:
        if (n % i) > 0:
            
            i += 1
            
        else:
            n  = n // i
            
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors[-1]
```

```python
print(' largest prime factor of the number 600851475143 is : ')
prime_factors(600851475143)
```

# 4
A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 × 99.

Find the largest palindrome made from the product of two 3-digit numbers.

```python
prod_num = []
palindrome_nums = []

for i in range (100, 1000):
    for j in range (100, 1000):
        prod_num.append(i * j)
for i in prod_num:        
    if list(str(i)) == list(reversed(str(i))):
        palindrome_nums.append(i)
print(max(palindrome_nums))           
```

# 5
2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?

```python
def gcd(a,b):
    """Compute the greatest common divisor of a and b"""
    while b > 0:
        a, b = b, a % b
    return a
    
def lcm(a, b):
    """Compute the lowest common multiple of a and b"""
    return a * b / gcd(a, b)

n = 1
for i in range(1, 21):
    n = lcm(n, i)
print(n)
```

a2 + b2 = c2
For example, 32 + 42 = 9 + 16 = 25 = 52.

There exists exactly one Pythagorean triplet for which a + b + c = 1000.
Find the product abc.

```python
for a in range(1,10):
    for b in range(1,10):
        for c in range(1,10):
            if a * a + b * b == c * c:
                print(a,b,c)
```

# 10
The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.

```python
import math
prime_list = []
num = 2 # prime numbers are greater than 1
while num < 10 :
   
       # check for factors
    for i in range(2,int(math.sqrt(num) + 1)):
        if (num % i) == 0: # is not a prime number

            break
    else:
        prime_list.append(num)
    num += 1       
print(sum(prime_list))
```
